```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv('groceries_dataset.csv')
```

```python
df.head()
```

```python
records = []
for i in range(0, 9835):
    records.append([str(df.values[i,u]) for u in range(0, 32)])
```

```python
rules = apriori(records, min_support=0.0022, min_confidence=0.20, min_lift=3, min_length = 4)
```

```python
# Store rules in result variable
results = list(rules)# See the items that were bought together with their support
results_list = []
for i in range(0, len(results)):
    results_list.append('RULE:\t' + str(results[i][0]) + '\nSUPPORT:\t' + str(results[i][1]))
```

```python
frequent_itemsets = apriori(data_processed, min_support=0.1, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
```

```python
rules['P(A, B)'] = rules['support']
rules['P(B|A)'] = rules['confidence']
rules['P(A, ~B)'] = rules.apply(lambda row: row['antecedent support'] - row['support'], axis=1)
rules['P(~B|A)'] = 1 - rules['confidence']
rules['P(B)'] = rules['consequent support']
rules['P(~B)'] = rules['consequent support'].apply(lambda x: 1 - x)
```

```python
rules['J-Measure'] = (
    rules['P(A, B)'] * np.log2(rules['P(B|A)'] / rules['P(B)']) +
    rules['P(A, ~B)'] * np.log2(rules['P(~B|A)'] / rules['P(~B)'])
)
```

```python
rules = rules.sort_values(by='J-Measure', ascending=False)
print("Top 10 rules with highest J-Measure:")
```
