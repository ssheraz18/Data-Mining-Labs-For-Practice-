```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Perceptron:
    def __init__(self, input_size, lr=1, epochs=1000):
        self.W = np.zeros(input_size+1)
        self.epochs = epochs
        self.lr = lr

    def activation_fn(self, x):
        return 1 if x >= 0 else 0

    def predict(self, x):
        return self.activation_fn(self.W.T.dot(x))

    def fit(self, X, d):
        for _ in range(self.epochs):
            for i in range(d.shape[0]):
                x = np.insert(X[i], 0, 1)
                self.W += self.lr * (d[i] - self.predict(x)) * x

def plot_decision_boundary(perceptron, X, d):
    x = np.linspace(0, 1.2)
    y = (perceptron.W[0] + perceptron.W[1]*x) / -perceptron.W[2]
    plt.plot(x, y, 'r')
    plt.scatter(X[:,0], X[:,1], c=d)
    plt.show()
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
d_or = np.array([0, 1, 1, 1])
d_and = np.array([0, 0, 0, 1])
perceptron_or = Perceptron(input_size=2)
perceptron_or.fit(X, d_or)
print("The weights for OR gate are:", perceptron_or.W)
plot_decision_boundary(perceptron_or, X, d_or)
perceptron_and = Perceptron(input_size=2)
perceptron_and.fit(X, d_and)
print("The weights for AND gate are:", perceptron_and.W)
plot_decision_boundary(perceptron_and, X, d_and)
```
